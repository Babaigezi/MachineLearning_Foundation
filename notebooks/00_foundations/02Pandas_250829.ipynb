```python
# EN: Sort by multiple keys and drop duplicate rows by subset.
# ZH: 多键排序与按子集去重。

df.sort_values(["exam_score", "age"], ascending=[False, True]).head(5) # ascending based on which column 
df.drop_duplicates(subset=["name"], keep="first")
```

```python
# EN: Category frequencies, unique counts per column.
# ZH: 类别频数与每列唯一值数量。

df["name"].value_counts()
df.nunique()
```

```python
# EN: Aggregation with multiple metrics; transform produces same-length series (feature engineering).
# ZH: 多指标聚合；transform 返回与原表等长（常用于特征工程）。

g = (df.groupby("name") # name col 里面每一个不同的名字是一组
       .agg(age_mean=("age", "mean"), # DataFrame.agg((col,"func"), axis=0, *args, **kwargs)[source]
            score_mean=("exam_score", "mean"),
            n=("exam_score", "count"))
       .reset_index()) # 还原成普通列
g.head()
# groupby(...).agg(...).reset_index()：三件套，得到规整的汇总表

# EN: z-score per column using transform (demo on exam_score).
# ZH: 使用 transform 计算列的 z-score（以 exam_score 为例）。
df["zscore"] = (df["exam_score"] - df["exam_score"].transform("mean")) / df["exam_score"].transform("std")
df.head()
# 如果用了groupby分组那一定要用transform把dataframe变回原来的size
```
